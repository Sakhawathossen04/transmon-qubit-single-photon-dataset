```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =========================
# Create figure folder
# =========================
FIG_DIR = "figures"
os.makedirs(FIG_DIR, exist_ok=True)

# =========================
# Load data
# Run notebook from same folder
# =========================
fields_summary = pd.read_csv("Reflection_fields_summary.csv")
scalars = pd.read_csv("Reflection_scalars.csv")
vectors = pd.read_csv("Reflection_vectors.csv")

mag_on = pd.read_csv("magnitude_onRes.csv")
phase_on = pd.read_csv("phase_onRes.csv")
mag_off = pd.read_csv("magnitude_offRes.csv")
phase_off = pd.read_csv("phase_offRes.csv")

print("Loaded files:")
print("fields_summary:", fields_summary.shape)
print("scalars:", scalars.shape)
print("vectors:", vectors.shape)
print("magnitude_onRes:", mag_on.shape)
print("phase_onRes:", phase_on.shape)
print("magnitude_offRes:", mag_off.shape)
print("phase_offRes:", phase_off.shape)

print("\nFields summary:")
display(fields_summary)

print("\nScalars:")
display(scalars)

print("\nVectors preview:")
display(vectors.head())

print("\nMagnitude on-resonance preview:")
display(mag_on.head())

print("\nPhase on-resonance preview:")
display(phase_on.head())


# =========================
# Extract axes
# =========================
freq = vectors["frequency"].dropna().to_numpy(dtype=float)
power = vectors["vna_power"].dropna().to_numpy(dtype=float)

mag_on_arr = mag_on.to_numpy(dtype=float)
phase_on_arr = phase_on.to_numpy(dtype=float)
mag_off_arr = mag_off.to_numpy(dtype=float)
phase_off_arr = phase_off.to_numpy(dtype=float)

print("\nAxis lengths:")
print("frequency points:", len(freq))
print("power points:", len(power))

print("\nMatrix shapes:")
print("mag_on:", mag_on_arr.shape)
print("phase_on:", phase_on_arr.shape)
print("mag_off:", mag_off_arr.shape)
print("phase_off:", phase_off_arr.shape)


# =========================
# Helper function
# =========================
def save_and_show(fig_name):
    plt.tight_layout()
    save_path = os.path.join(FIG_DIR, fig_name)
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    print(f"Saved: {save_path}")
    plt.show()
    plt.close()


# =========================
# Build tidy dataframe
# =========================
rows = []
for i, f in enumerate(freq):
    for j, p in enumerate(power):
        rows.append({
            "frequency_Hz": f,
            "vna_power_dBm": p,
            "mag_onRes": mag_on_arr[i, j],
            "phase_onRes": phase_on_arr[i, j],
            "mag_offRes": mag_off_arr[i, j],
            "phase_offRes": phase_off_arr[i, j],
            "mag_diff_on_minus_off": mag_on_arr[i, j] - mag_off_arr[i, j],
            "phase_diff_on_minus_off": phase_on_arr[i, j] - phase_off_arr[i, j],
        })

df = pd.DataFrame(rows)

print("\nTidy dataframe shape:", df.shape)
print("\nTidy dataframe preview:")
display(df.head())


# =========================
# Figure 1: On-resonance magnitude heatmap
# =========================
plt.figure(figsize=(8, 6))
plt.imshow(
    mag_on_arr,
    aspect="auto",
    extent=[power.min(), power.max(), freq.min(), freq.max()],
    origin="lower"
)
plt.xlabel("VNA Power (dBm)")
plt.ylabel("Frequency (Hz)")
plt.title("Reflection Magnitude (On Resonance)")
plt.colorbar(label="Magnitude")
save_and_show("fig1_mag_onRes_heatmap.png")


# =========================
# Figure 2: Off-resonance magnitude heatmap
# =========================
plt.figure(figsize=(8, 6))
plt.imshow(
    mag_off_arr,
    aspect="auto",
    extent=[power.min(), power.max(), freq.min(), freq.max()],
    origin="lower"
)
plt.xlabel("VNA Power (dBm)")
plt.ylabel("Frequency (Hz)")
plt.title("Reflection Magnitude (Off Resonance)")
plt.colorbar(label="Magnitude")
save_and_show("fig2_mag_offRes_heatmap.png")


# =========================
# Figure 3: On-resonance phase heatmap
# =========================
plt.figure(figsize=(8, 6))
plt.imshow(
    phase_on_arr,
    aspect="auto",
    extent=[power.min(), power.max(), freq.min(), freq.max()],
    origin="lower"
)
plt.xlabel("VNA Power (dBm)")
plt.ylabel("Frequency (Hz)")
plt.title("Reflection Phase (On Resonance)")
plt.colorbar(label="Phase (rad)")
save_and_show("fig3_phase_onRes_heatmap.png")


# =========================
# Figure 4: Off-resonance phase heatmap
# =========================
plt.figure(figsize=(8, 6))
plt.imshow(
    phase_off_arr,
    aspect="auto",
    extent=[power.min(), power.max(), freq.min(), freq.max()],
    origin="lower"
)
plt.xlabel("VNA Power (dBm)")
plt.ylabel("Frequency (Hz)")
plt.title("Reflection Phase (Off Resonance)")
plt.colorbar(label="Phase (rad)")
save_and_show("fig4_phase_offRes_heatmap.png")


# =========================
# Figure 5: Magnitude difference heatmap
# =========================
mag_diff = mag_on_arr - mag_off_arr

plt.figure(figsize=(8, 6))
plt.imshow(
    mag_diff,
    aspect="auto",
    extent=[power.min(), power.max(), freq.min(), freq.max()],
    origin="lower"
)
plt.xlabel("VNA Power (dBm)")
plt.ylabel("Frequency (Hz)")
plt.title("Magnitude Difference: On - Off Resonance")
plt.colorbar(label="Magnitude Difference")
save_and_show("fig5_mag_difference_heatmap.png")


# =========================
# Figure 6: Phase difference heatmap
# =========================
phase_diff = phase_on_arr - phase_off_arr

plt.figure(figsize=(8, 6))
plt.imshow(
    phase_diff,
    aspect="auto",
    extent=[power.min(), power.max(), freq.min(), freq.max()],
    origin="lower"
)
plt.xlabel("VNA Power (dBm)")
plt.ylabel("Frequency (Hz)")
plt.title("Phase Difference: On - Off Resonance")
plt.colorbar(label="Phase Difference")
save_and_show("fig6_phase_difference_heatmap.png")


# =========================
# Figure 7: Line cut at middle power (magnitude)
# =========================
mid_idx = len(power) // 2

plt.figure(figsize=(10, 5))
plt.plot(freq, mag_on_arr[:, mid_idx], label="On resonance")
plt.plot(freq, mag_off_arr[:, mid_idx], label="Off resonance")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title(f"Magnitude Line Cut at VNA Power = {power[mid_idx]:.1f} dBm")
plt.legend()
save_and_show("fig7_magnitude_linecut_midpower.png")


# =========================
# Figure 8: Line cut at middle power (phase)
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq, phase_on_arr[:, mid_idx], label="On resonance")
plt.plot(freq, phase_off_arr[:, mid_idx], label="Off resonance")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (rad)")
plt.title(f"Phase Line Cut at VNA Power = {power[mid_idx]:.1f} dBm")
plt.legend()
save_and_show("fig8_phase_linecut_midpower.png")


# =========================
# Figure 9: Several power cuts for on-res magnitude
# =========================
selected_power_indices = [0, len(power)//4, len(power)//2, 3*len(power)//4, len(power)-1]

plt.figure(figsize=(10, 6))
for idx in selected_power_indices:
    plt.plot(freq, mag_on_arr[:, idx], label=f"{power[idx]:.1f} dBm")

plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("On-Resonance Magnitude at Selected Power Levels")
plt.legend()
save_and_show("fig9_mag_onRes_selected_powercuts.png")


# =========================
# Figure 10: Several power cuts for off-res magnitude
# =========================
plt.figure(figsize=(10, 6))
for idx in selected_power_indices:
    plt.plot(freq, mag_off_arr[:, idx], label=f"{power[idx]:.1f} dBm")

plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Off-Resonance Magnitude at Selected Power Levels")
plt.legend()
save_and_show("fig10_mag_offRes_selected_powercuts.png")


# =========================
# Figure 11: Several power cuts for on-res phase
# =========================
plt.figure(figsize=(10, 6))
for idx in selected_power_indices:
    plt.plot(freq, phase_on_arr[:, idx], label=f"{power[idx]:.1f} dBm")

plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (rad)")
plt.title("On-Resonance Phase at Selected Power Levels")
plt.legend()
save_and_show("fig11_phase_onRes_selected_powercuts.png")


# =========================
# Figure 12: Several power cuts for off-res phase
# =========================
plt.figure(figsize=(10, 6))
for idx in selected_power_indices:
    plt.plot(freq, phase_off_arr[:, idx], label=f"{power[idx]:.1f} dBm")

plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (rad)")
plt.title("Off-Resonance Phase at Selected Power Levels")
plt.legend()
save_and_show("fig12_phase_offRes_selected_powercuts.png")


# =========================
# Figure 13: Mean magnitude vs power
# =========================
mag_on_mean_vs_power = mag_on_arr.mean(axis=0)
mag_off_mean_vs_power = mag_off_arr.mean(axis=0)

plt.figure(figsize=(8, 5))
plt.plot(power, mag_on_mean_vs_power, marker="o", label="On resonance")
plt.plot(power, mag_off_mean_vs_power, marker="o", label="Off resonance")
plt.xlabel("VNA Power (dBm)")
plt.ylabel("Mean Magnitude")
plt.title("Mean Magnitude vs VNA Power")
plt.legend()
save_and_show("fig13_mean_magnitude_vs_power.png")


# =========================
# Figure 14: Mean phase vs power
# =========================
phase_on_mean_vs_power = phase_on_arr.mean(axis=0)
phase_off_mean_vs_power = phase_off_arr.mean(axis=0)

plt.figure(figsize=(8, 5))
plt.plot(power, phase_on_mean_vs_power, marker="o", label="On resonance")
plt.plot(power, phase_off_mean_vs_power, marker="o", label="Off resonance")
plt.xlabel("VNA Power (dBm)")
plt.ylabel("Mean Phase (rad)")
plt.title("Mean Phase vs VNA Power")
plt.legend()
save_and_show("fig14_mean_phase_vs_power.png")


# =========================
# Figure 15: Minimum on-res magnitude frequency vs power
# =========================
min_freqs = []
min_vals = []

for j in range(len(power)):
    idx_min = np.argmin(mag_on_arr[:, j])
    min_freqs.append(freq[idx_min])
    min_vals.append(mag_on_arr[idx_min, j])

plt.figure(figsize=(8, 5))
plt.plot(power, min_freqs, marker="o")
plt.xlabel("VNA Power (dBm)")
plt.ylabel("Frequency of Minimum On-Res Magnitude (Hz)")
plt.title("Estimated Resonance Frequency vs Power")
save_and_show("fig15_resonance_frequency_vs_power.png")


# =========================
# Figure 16: Minimum on-res magnitude value vs power
# =========================
plt.figure(figsize=(8, 5))
plt.plot(power, min_vals, marker="o")
plt.xlabel("VNA Power (dBm)")
plt.ylabel("Minimum On-Res Magnitude")
plt.title("Minimum On-Res Magnitude vs Power")
save_and_show("fig16_minimum_magnitude_vs_power.png")


print("\nDone. All figures are shown and saved in the 'figures' folder.")
```
