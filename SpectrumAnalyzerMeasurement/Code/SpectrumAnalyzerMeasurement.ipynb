```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =========================
# Create figure folder
# =========================
FIG_DIR = "figures"
os.makedirs(FIG_DIR, exist_ok=True)

# =========================
# Load data
# Run notebook from same folder
# =========================
fields_summary = pd.read_csv("SpectrumAnalyzerMeasurement_fields_summary.csv")
scalars = pd.read_csv("SpectrumAnalyzerMeasurement_scalars.csv")
vectors = pd.read_csv("SpectrumAnalyzerMeasurement_vectors.csv")
df = pd.read_csv("SpectrumAnalyzerMeasurement_long_dataframe.csv")

print("Loaded files:")
print("fields_summary:", fields_summary.shape)
print("scalars:", scalars.shape)
print("vectors:", vectors.shape)
print("long dataframe:", df.shape)

print("\nFields summary:")
display(fields_summary)

print("\nScalars:")
display(scalars)

print("\nVectors preview:")
display(vectors.head())

print("\nLong dataframe preview:")
display(df.head())


# =========================
# Extract arrays
# =========================
freq = df["frequency"].to_numpy(dtype=float)

sa_noise = df["SANoiseTrace"].to_numpy(dtype=float)
hemt_noise = df["HEMTNoiseTrace"].to_numpy(dtype=float)
jpa_noise = df["JPANoiseTrace"].to_numpy(dtype=float)

hemt_vna = df["HEMT_VNATrace"].to_numpy(dtype=float)
jpa_vna = df["JPA_VNATrace"].to_numpy(dtype=float)

hemt_lo = df["HEMT_LOTrace"].to_numpy(dtype=float)
jpa_lo = df["JPA_LOTrace"].to_numpy(dtype=float)

# scalar metadata as dict
meta = dict(zip(scalars["field"], scalars["value"]))

qubit_f = float(meta.get("qubit_f", np.nan))
sa_noise_level = float(meta.get("SANoiseLevel", np.nan))
hemt_noise_level = float(meta.get("HEMTNoiseLevel", np.nan))
jpa_noise_level = float(meta.get("JPANoiseLevel", np.nan))
hemt_vna_peak_level = float(meta.get("HEMT_VNAPeakLevel", np.nan))
jpa_vna_peak_level = float(meta.get("JPA_VNAPeakLevel", np.nan))
hemt_lo_peak_level = float(meta.get("HEMT_LOPeakLevel", np.nan))
jpa_lo_peak_level = float(meta.get("JPA_LOPeakLevel", np.nan))

print("\nExtracted metadata:")
for k, v in meta.items():
    print(f"{k}: {v}")


# =========================
# Helper functions
# =========================
def save_and_show(fig_name):
    plt.tight_layout()
    save_path = os.path.join(FIG_DIR, fig_name)
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    print(f"Saved: {save_path}")
    plt.show()
    plt.close()

def to_db(x):
    x = np.asarray(x, dtype=float)
    return 10 * np.log10(np.maximum(x, 1e-30))

def normalize_trace(x):
    x = np.asarray(x, dtype=float)
    xmin = np.min(x)
    xmax = np.max(x)
    if xmax == xmin:
        return np.zeros_like(x)
    return (x - xmin) / (xmax - xmin)

def peak_info(x, y):
    idx = np.argmax(y)
    return idx, x[idx], y[idx]


# =========================
# Peak detection
# =========================
idx_hemt_vna, freq_hemt_vna_peak, val_hemt_vna_peak = peak_info(freq, hemt_vna)
idx_jpa_vna, freq_jpa_vna_peak, val_jpa_vna_peak = peak_info(freq, jpa_vna)

idx_hemt_lo, freq_hemt_lo_peak, val_hemt_lo_peak = peak_info(freq, hemt_lo)
idx_jpa_lo, freq_jpa_lo_peak, val_jpa_lo_peak = peak_info(freq, jpa_lo)

idx_sa_noise, freq_sa_noise_peak, val_sa_noise_peak = peak_info(freq, sa_noise)
idx_hemt_noise, freq_hemt_noise_peak, val_hemt_noise_peak = peak_info(freq, hemt_noise)
idx_jpa_noise, freq_jpa_noise_peak, val_jpa_noise_peak = peak_info(freq, jpa_noise)

print("\nDetected peak frequencies:")
print(f"HEMT VNA peak: {freq_hemt_vna_peak:.6e} Hz, value = {val_hemt_vna_peak:.6e}")
print(f"JPA VNA peak:  {freq_jpa_vna_peak:.6e} Hz, value = {val_jpa_vna_peak:.6e}")
print(f"HEMT LO peak:  {freq_hemt_lo_peak:.6e} Hz, value = {val_hemt_lo_peak:.6e}")
print(f"JPA LO peak:   {freq_jpa_lo_peak:.6e} Hz, value = {val_jpa_lo_peak:.6e}")
print(f"SA noise peak:   {freq_sa_noise_peak:.6e} Hz, value = {val_sa_noise_peak:.6e}")
print(f"HEMT noise peak: {freq_hemt_noise_peak:.6e} Hz, value = {val_hemt_noise_peak:.6e}")
print(f"JPA noise peak:  {freq_jpa_noise_peak:.6e} Hz, value = {val_jpa_noise_peak:.6e}")


# =========================
# Build helper dataframe
# =========================
analysis_df = pd.DataFrame({
    "frequency_Hz": freq,
    "frequency_offset_Hz": freq - qubit_f,
    "SANoiseTrace": sa_noise,
    "HEMTNoiseTrace": hemt_noise,
    "JPANoiseTrace": jpa_noise,
    "HEMT_VNATrace": hemt_vna,
    "JPA_VNATrace": jpa_vna,
    "HEMT_LOTrace": hemt_lo,
    "JPA_LOTrace": jpa_lo,
    "SANoiseTrace_dB": to_db(sa_noise),
    "HEMTNoiseTrace_dB": to_db(hemt_noise),
    "JPANoiseTrace_dB": to_db(jpa_noise),
    "HEMT_VNATrace_dB": to_db(hemt_vna),
    "JPA_VNATrace_dB": to_db(jpa_vna),
    "HEMT_LOTrace_dB": to_db(hemt_lo),
    "JPA_LOTrace_dB": to_db(jpa_lo),
})

print("\nAnalysis dataframe preview:")
display(analysis_df.head())


# =========================
# Figure 1: All traces (linear)
# =========================
plt.figure(figsize=(10, 6))
plt.plot(freq, sa_noise, label="SA Noise")
plt.plot(freq, hemt_noise, label="HEMT Noise")
plt.plot(freq, jpa_noise, label="JPA Noise")
plt.plot(freq, hemt_vna, label="HEMT VNA")
plt.plot(freq, jpa_vna, label="JPA VNA")
plt.plot(freq, hemt_lo, label="HEMT LO")
plt.plot(freq, jpa_lo, label="JPA LO")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Trace value")
plt.title("All Spectrum Analyzer Traces (Linear Scale)")
plt.legend()
save_and_show("fig1_all_traces_linear.png")


# =========================
# Figure 2: All traces (dB)
# =========================
plt.figure(figsize=(10, 6))
plt.plot(freq, to_db(sa_noise), label="SA Noise")
plt.plot(freq, to_db(hemt_noise), label="HEMT Noise")
plt.plot(freq, to_db(jpa_noise), label="JPA Noise")
plt.plot(freq, to_db(hemt_vna), label="HEMT VNA")
plt.plot(freq, to_db(jpa_vna), label="JPA VNA")
plt.plot(freq, to_db(hemt_lo), label="HEMT LO")
plt.plot(freq, to_db(jpa_lo), label="JPA LO")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power-like value (dB)")
plt.title("All Spectrum Analyzer Traces (dB Scale)")
plt.legend()
save_and_show("fig2_all_traces_db.png")


# =========================
# Figure 3: Noise traces comparison
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq, sa_noise, label="SA Noise")
plt.plot(freq, hemt_noise, label="HEMT Noise")
plt.plot(freq, jpa_noise, label="JPA Noise")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Trace value")
plt.title("Noise Trace Comparison")
plt.legend()
save_and_show("fig3_noise_trace_comparison.png")


# =========================
# Figure 4: Noise traces comparison in dB
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq, to_db(sa_noise), label="SA Noise")
plt.plot(freq, to_db(hemt_noise), label="HEMT Noise")
plt.plot(freq, to_db(jpa_noise), label="JPA Noise")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power-like value (dB)")
plt.title("Noise Trace Comparison (dB)")
plt.legend()
save_and_show("fig4_noise_trace_comparison_db.png")


# =========================
# Figure 5: VNA traces comparison
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq, hemt_vna, label="HEMT VNA")
plt.plot(freq, jpa_vna, label="JPA VNA")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Trace value")
plt.title("VNA Trace Comparison")
plt.legend()
save_and_show("fig5_vna_trace_comparison.png")


# =========================
# Figure 6: LO traces comparison
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq, hemt_lo, label="HEMT LO")
plt.plot(freq, jpa_lo, label="JPA LO")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Trace value")
plt.title("LO Trace Comparison")
plt.legend()
save_and_show("fig6_lo_trace_comparison.png")


# =========================
# Figure 7: Normalized VNA traces
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq, normalize_trace(hemt_vna), label="HEMT VNA (normalized)")
plt.plot(freq, normalize_trace(jpa_vna), label="JPA VNA (normalized)")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Normalized amplitude")
plt.title("Normalized VNA Trace Comparison")
plt.legend()
save_and_show("fig7_normalized_vna_traces.png")


# =========================
# Figure 8: Normalized LO traces
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq, normalize_trace(hemt_lo), label="HEMT LO (normalized)")
plt.plot(freq, normalize_trace(jpa_lo), label="JPA LO (normalized)")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Normalized amplitude")
plt.title("Normalized LO Trace Comparison")
plt.legend()
save_and_show("fig8_normalized_lo_traces.png")


# =========================
# Figure 9: Frequency offset plot around qubit frequency
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq - qubit_f, hemt_vna, label="HEMT VNA")
plt.plot(freq - qubit_f, jpa_vna, label="JPA VNA")
plt.plot(freq - qubit_f, hemt_lo, label="HEMT LO")
plt.plot(freq - qubit_f, jpa_lo, label="JPA LO")
plt.axvline(0, linestyle="--", label="Qubit f offset")
plt.xlabel("Frequency offset from qubit_f (Hz)")
plt.ylabel("Trace value")
plt.title("Spectral Traces Centered Around Qubit Frequency")
plt.legend()
save_and_show("fig9_frequency_offset_centered.png")


# =========================
# Figure 10: Peak markers for VNA traces
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq, hemt_vna, label="HEMT VNA")
plt.plot(freq, jpa_vna, label="JPA VNA")
plt.scatter([freq_hemt_vna_peak], [val_hemt_vna_peak], s=80, label="HEMT VNA peak")
plt.scatter([freq_jpa_vna_peak], [val_jpa_vna_peak], s=80, label="JPA VNA peak")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Trace value")
plt.title("VNA Peaks")
plt.legend()
save_and_show("fig10_vna_peaks.png")


# =========================
# Figure 11: Peak markers for LO traces
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq, hemt_lo, label="HEMT LO")
plt.plot(freq, jpa_lo, label="JPA LO")
plt.scatter([freq_hemt_lo_peak], [val_hemt_lo_peak], s=80, label="HEMT LO peak")
plt.scatter([freq_jpa_lo_peak], [val_jpa_lo_peak], s=80, label="JPA LO peak")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Trace value")
plt.title("LO Peaks")
plt.legend()
save_and_show("fig11_lo_peaks.png")


# =========================
# Figure 12: Peak level bar plot
# =========================
peak_names = ["HEMT_VNA", "JPA_VNA", "HEMT_LO", "JPA_LO"]
peak_values = [hemt_vna_peak_level, jpa_vna_peak_level, hemt_lo_peak_level, jpa_lo_peak_level]

plt.figure(figsize=(8, 5))
plt.bar(peak_names, peak_values)
plt.ylabel("Peak level")
plt.title("Peak Level Comparison")
save_and_show("fig12_peak_level_barplot.png")


# =========================
# Figure 13: Noise level bar plot
# =========================
noise_names = ["SA", "HEMT", "JPA"]
noise_values = [sa_noise_level, hemt_noise_level, jpa_noise_level]

plt.figure(figsize=(8, 5))
plt.bar(noise_names, noise_values)
plt.ylabel("Noise level")
plt.title("Noise Level Comparison")
save_and_show("fig13_noise_level_barplot.png")


# =========================
# Figure 14: JPA/HEMT gain-like ratio for VNA and LO
# =========================
vna_ratio = jpa_vna / np.maximum(hemt_vna, 1e-30)
lo_ratio = jpa_lo / np.maximum(hemt_lo, 1e-30)

plt.figure(figsize=(10, 5))
plt.plot(freq, vna_ratio, label="JPA / HEMT (VNA)")
plt.plot(freq, lo_ratio, label="JPA / HEMT (LO)")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Ratio")
plt.title("JPA-to-HEMT Ratio")
plt.legend()
save_and_show("fig14_jpa_hemt_ratio.png")


# =========================
# Figure 15: JPA/HEMT ratio in dB
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq, to_db(vna_ratio), label="JPA / HEMT (VNA) [dB]")
plt.plot(freq, to_db(lo_ratio), label="JPA / HEMT (LO) [dB]")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Ratio (dB)")
plt.title("JPA-to-HEMT Ratio in dB")
plt.legend()
save_and_show("fig15_jpa_hemt_ratio_db.png")


# =========================
# Figure 16: Zoom around qubit frequency
# =========================
window_hz = 2.5e5
mask = np.abs(freq - qubit_f) <= window_hz

plt.figure(figsize=(10, 5))
plt.plot(freq[mask], hemt_vna[mask], label="HEMT VNA")
plt.plot(freq[mask], jpa_vna[mask], label="JPA VNA")
plt.plot(freq[mask], hemt_lo[mask], label="HEMT LO")
plt.plot(freq[mask], jpa_lo[mask], label="JPA LO")
plt.axvline(qubit_f, linestyle="--", label="Qubit f")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Trace value")
plt.title("Zoom Around Qubit Frequency")
plt.legend()
save_and_show("fig16_zoom_around_qubit_frequency.png")


# =========================
# Figure 17: Histogram of trace amplitudes
# =========================
plt.figure(figsize=(10, 5))
plt.hist(sa_noise, bins=50, alpha=0.4, label="SA Noise")
plt.hist(hemt_noise, bins=50, alpha=0.4, label="HEMT Noise")
plt.hist(jpa_noise, bins=50, alpha=0.4, label="JPA Noise")
plt.xlabel("Trace value")
plt.ylabel("Count")
plt.title("Histogram of Noise Trace Amplitudes")
plt.legend()
save_and_show("fig17_hist_noise_traces.png")


# =========================
# Figure 18: Histogram of VNA/LO amplitudes
# =========================
plt.figure(figsize=(10, 5))
plt.hist(hemt_vna, bins=50, alpha=0.4, label="HEMT VNA")
plt.hist(jpa_vna, bins=50, alpha=0.4, label="JPA VNA")
plt.hist(hemt_lo, bins=50, alpha=0.4, label="HEMT LO")
plt.hist(jpa_lo, bins=50, alpha=0.4, label="JPA LO")
plt.xlabel("Trace value")
plt.ylabel("Count")
plt.title("Histogram of VNA and LO Trace Amplitudes")
plt.legend()
save_and_show("fig18_hist_vna_lo_traces.png")


# =========================
# Figure 19: dB traces around qubit frequency
# =========================
plt.figure(figsize=(10, 5))
plt.plot(freq - qubit_f, to_db(hemt_vna), label="HEMT VNA")
plt.plot(freq - qubit_f, to_db(jpa_vna), label="JPA VNA")
plt.plot(freq - qubit_f, to_db(hemt_lo), label="HEMT LO")
plt.plot(freq - qubit_f, to_db(jpa_lo), label="JPA LO")
plt.axvline(0, linestyle="--", label="Qubit f offset")
plt.xlabel("Frequency offset from qubit_f (Hz)")
plt.ylabel("Power-like value (dB)")
plt.title("dB Traces Centered Around Qubit Frequency")
plt.legend()
save_and_show("fig19_db_traces_centered.png")


# =========================
# Figure 20: Table of detected peaks
# =========================
peak_table = pd.DataFrame({
    "trace": [
        "HEMT_VNATrace", "JPA_VNATrace",
        "HEMT_LOTrace", "JPA_LOTrace",
        "SANoiseTrace", "HEMTNoiseTrace", "JPANoiseTrace"
    ],
    "peak_frequency_Hz": [
        freq_hemt_vna_peak, freq_jpa_vna_peak,
        freq_hemt_lo_peak, freq_jpa_lo_peak,
        freq_sa_noise_peak, freq_hemt_noise_peak, freq_jpa_noise_peak
    ],
    "peak_value": [
        val_hemt_vna_peak, val_jpa_vna_peak,
        val_hemt_lo_peak, val_jpa_lo_peak,
        val_sa_noise_peak, val_hemt_noise_peak, val_jpa_noise_peak
    ]
})

print("\nDetected peak table:")
display(peak_table)

print("\nDone. All figures are shown and saved in the 'figures' folder.")
```
